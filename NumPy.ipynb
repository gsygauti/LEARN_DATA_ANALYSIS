# NUMPY ARRAY INDEXING - 
array indexing is accessing array elements. In Numpy, access an array element using the index number. The index[0] = element 1 , index[1] = element 2 and so on...

```python
## access elements from a 1D Array.
import numpy as np

n = np.array([10, 20, 30, 40, 50])
print(n,n.ndim,"-D array")
print(n[0]) ## access the 1st element.
print(n[3]) ## access the 4th element.

k = int(input("enter the which (1-5)position element you want to access: "))
print(n[k-1])  ## access the kth element present at inndex[k-1].
```

```python
## access elements from a 2D Array.
import numpy as np

n = np.array([[1,2,7],[3,4,8],[5,6,9]])
print(n,n.ndim,"-D array")
print(n[0,0],"= 1")       ## n[row,col] both starting from [0,0].
print(n[1,2],"= 4")
print(n[2,2],"= 8")
```

```python
## access elements from a 3D Array.(3D - consist of 2D arrays).
import numpy as np

n = np.array([[[1,2,9],      ## x = 0 ; y= 0-2 ; z = 0-2
               [3,4,10]],
              [[5,6,11],      ## x = 1 ; y= 0-2 ; z= 0-2
               [7,8,12]]])

print(n,n.ndim,"-D array") ## n[row,col] both starting from [0,0,0] , ending[1,1,2].
print(n[0,0,0],"= 1")
print(n[0,1,1],"= 4")
print(n[1,0,0],"= 5")
print(n[1,1,0],"= 7")
print(n[1,1,2],"= 12")
print(n[1,0,1],"= 6")
```

```python
## access elements from the last with Negtive indexing.
import numpy as np

       ## x = -2 ; y,z= ([-3]-[-1])       ## x = -1 ; y,z= ([-3]-[-1])
n = np.array([ [ [1,2,9],[3,4,10] ] , [ [5,6,11],[7,8,12] ] ])

print(n, n.ndim,"-D array") ## n[row,col] both starting from [-2,-2,-3] , ending [-1,-1,-1]
print(n[-2,-2,-3],"= 1")
print(n[-2,-1,-2],"= 4")
print(n[-1,-2,-3],"= 5")
print(n[-1,-1,-3],"= 7")
print(n[-1,-1,-1],"= 12")
print(n[-1,-2,-2],"= 6")
print(n[-2,-2,-1],"= 9")
```

# NUMPY SLICING IN (1-D) ARRAY- 
slicing the array in range from start to end is what we call Array slicing. Slice(range) is passed in place of index. syantax: - arr[ start : end ]
* here, start is (include) because starting from there but the end is (excluded) becoz starting from index 0 so it will go from 0 -(n-1) positions.
* SLICING with step - arr[ start : end : step ].


```python
## Q.1 Slicing from index 2 to 5 (5exculed)

import numpy as np

n = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])

print(n[2: 5])
```

```python
## Q.2 Slicing from index 5 to last

n = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
k = len(n)

print(n[5: k+1])
```

```python
## Q.3 Slicing from starting to 5 (5 exculed) 
n = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])

print(n[ : 5])
```

```python
## Q.4 Sliing from with jump/step is 2.

n = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])

print(n[: : 2])  ## or n[0:10:2] starting from odd position
print(n[1:10:2])           ## starting from even position.
```

# NUMPY SLICING IN (2-D) ARRAY-
syntax - arr[(0/1) , start : end ]
* 2D slicing syntax : 
array[ row_start : row_end , col_start : col_end ]

```python
## Q.5 slicing from index 2 to 5(excluding 5)

n = np.array([ [5, 10, 15, 20, 25],   
              [30, 35, 40, 45, 50] ])

print(f"slicing in {n.ndim}D array\n")

## when print one row only.

print(n[0, 2:5])   ## n [(0/1) , start : end ]
print(n[1, 2:5 ]) 
```

```python
## now let's combine both in one go.

n = np.array([ [5, 10, 15, 20, 25],   
              [30, 35, 40, 45, 50] ])

print(f"slicing in {n.ndim}D array\n")

## when print multiple row.

print(n[0:2, 2:5])       ## n[row_start:row_end, col_start:col_end] --> n[row 0 : row 2 , col 2 : col 5].
```

# JOIN NUMPY ARRAYS- 
we can join two or more arrays in a single new array using the following:
1. concatenate() method: joining along with axis, unlike keys in SQL.
2. stack methods: joining along a new axis, unlike keys in SQL. ( stack(), hstack(), vstack(), dstack(), column_stack() )

```python
## by using 1st method.

import numpy as np

n1 = np.array([1,3,5,7,9])
n2 = np.array([2,4,6,8,10])

result = np.concatenate((n1,n2)) ### use numpy.concatenate( (val1, val2) );   note - (val1,val2) must be in a new bracket.
print(result)
```

```python
## make result arr a natural number sequence.

import numpy as np

n1 = np.array([1,3,5,7,9])
n2 = np.array([2,4,6,8,10])

result = np.concatenate((n1,n2)) ### use numpy.concatenate( (val1, val2) );   note - (val1,val2) must be in a new bracket.
result.sort()
print(result)
```

##  NOTE - IN all stack() methods - 
* all input arrays must have the same shape.  otherwise throug:  ValueError
* the stack() make two array in 2-D and by defult takes (0,0) indexing.
* the dstack() make the transpose of result of the stack() and vstack(). it means depth(height) convert len(arr) into height(arr) or M[col]-> M[row]
* the dstack() also make two arr in 3-D matrix.

```python
## by using 2nd method. 

## stack() it will give the array in 2-D arr.

import numpy as np

n1 = np.array([1,3,5,7,9])
n2 = np.array([2,4,6,8,10])

result = np.stack((n1, n2))  ## don't mention axis so by defult(0,0)
print(result)
```

```python
## hstack() it used in stack along rows using the hstack(). in numpy. 

import numpy as np

n1 = np.array([1,3,5,7,9])
n2 = np.array([2,4,6,8,10])

result = np.hstack((n1, n2))  # join along rows.
print(result)
```

```python
## vstack() it used in stack along columns using the vstack(). in numpy. 

import numpy as np

n1 = np.array([1,3,5,7,9])
n2 = np.array([2,4,6,8,10])

result = np.vstack((n1, n2))  # join along columns.
print(result)
```

```python
## dstack() it used in stack along depth(height)  using the dstack(). in numpy. it will give 3-D arr.

import numpy as np

n1 = np.array([1,3,5,7,9])
n2 = np.array([2,4,6,8,10])

result = np.dstack((n1, n2))  ## stack along rows of both arr.
print(result)
```

```python
## columns_stack() it used in stack according to  columns using the column_stack(). in numpy. 

import numpy as np

n1 = np.array([1,3,5,7,9])
n2 = np.array([2,4,6,8,10])

result = np.column_stack((n1, n2))  ## stack along colunns of both arr.
print(result)
```

# NUMPY ARRAYS INTERSECTION- 
for intersection of two arrays, use the ( intersect1d( ) ) method in Numpy. intersection means common element in both array.

```python
## Q.1 find the intersection between two arrays.

import numpy as np

n1 = np.array([1, 6, 3, 8, 5])
n2 = np.array([6, 7, 8, 9, 10])

result = np.intersect1d(n1,n2)

print("the first array: ", n1)
print("the second array: ", n2)
print("intersection of two arr: ", result)
```

# STATISTICAL OPERATIONS ON NUMPY ARRAYS-
We can perform statistical operation like ( mean, median and standard deviation )on numpy arrays.
1. Mean: The average value of a dataset, calculated by summing all the numbers and dividing by the total count of values.
2. Median: The middle value in a sorted list of numbers. If the list has an even number of elements, it’s the average of the two middle values.
3. Standard Deviation (SD): A measure of how spread out the numbers are from the mean; a higher SD means more variation in the data.

```python
## Mean is the avg of the given values. to find mean used ( np.mean( arr) ) method. 

import numpy as np

n = np.array([5, 7, 8, 9, 10, 11])

result = np.mean(n)
print(result)
```

```python
## Median is the middle value of the given arr. to find median used ( np.medain( arr) ) method. 

import numpy as np

n = np.array([10, 20, 30, 40, 50])

result = np.median(n)

print(result)
```

```python
## standard Deviation is a measure of how spread out the number are from the mean of given values. to find mean used ( np.std( arr) ) method. 

import numpy as np

n = np.array([10, 20, 30, 40, 50])

result = np.std(n)
print(result)
```

# NUMPY LCM & HCF(GCD)-
To find LCM AND HCF of two numbers in Numpy. by using numpy.lcm( ) for LCM and numpy.gcd( ) for HCF(GCD).

```python
## Example of LCM

import numpy as np

a = int(input("enter the number1: "))
b = int(input("enter the number2: "))

result = np.lcm(a,b)   ## give indiviusal number.
print("the LCM of a, b: ",result)
```

```python
## Exmaple of LCM when more then 2 arrguments.

import numpy as np

a = int(input("Enter the number1: "))
b = int(input("Enter the number2: "))
c = int(input("Enter the number3: "))

# Compute LCM of a and b, then compute LCM of that result with c. becoz np.lcm take only 2 arrgument.

result = np.lcm.reduce([a, b, c])
print("LCM of the numbers is:", result)
```

```python
## Example of HCF(GCD)
import numpy as np

a = int(input("enter the number1: "))
b = int(input("enter the number2: "))

result = np.gcd(a,b)   ## give indiviusal number.
print("the HCF(GCD) of a, b: ",result)
```

```python
## Exmaple of HCF(GCD) when more then 2 arrguments.
import numpy as np

a = int(input("Enter the number1: "))
b = int(input("Enter the number2: "))
c = int(input("Enter the number3: "))

# Compute LCM of a and b, then compute LCM of that result with c. becoz np.lcm take only 2 arrgument.

result = np.gcd.reduce([a, b, c])

print("LCM of the numbers is:", result)
```
