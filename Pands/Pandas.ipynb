## Q2. ACCESS A GROUP OF ROWS OR COLUMNS IN A PANDAS DATAFRAME-

```python
import pandas as pd

# dataset
data = {
    'STUDENT':["AMIT", "JOHN", "BRIJESH", "SANDEEP", "UJJWAL"],
     'RANK': [1, 4, 3, 5, 2],
      'MARKS': [95, 70, 80, 60, 90]
}

## create a dataframe using DataFrame() method with index.
df = pd.DataFrame(data)

## display the records
print("Studnet records:\n",df)

## access the value in the student col corresponding to row label.      used df.loc(row,column)
print("\n val1 name of student of row index[0] = ",df.loc[ 0,'STUDENT'])
print("\n val2 rank of student of row index[0] = ", df.loc[ 0,'RANK'])
print("\n val3 marks of student of row index[0] = ", df.loc[ 0, 'MARKS'])
print("\n val4 rank of student of row index[1] =  ", df.loc[ 1,'RANK'])
print("\n val5 marks of student of row index[2] = ", df.loc[ 2, 'MARKS'])

```

# #5. JOIN PANDAS DATAFRAME-
in python, we can easily join Pandas Dataframe using the **join( ) method. this will join the columns of the two different Dataframes**.

```python
import pandas as pd

data1 = {
    "Id": ["so1", "so2", "so3", "so4", "so5"],
    "Student": ["Amit", "Brijesh", "Vansh", "Sandeep", "Ujjwal"],
    "Roll_no": [101, 102, 103, 104, 105]
}

data2 = {
    "Rank": [1, 4, 3, 5, 2],
    "marks": [95, 70, 80, 60,90]
}

## create dataframes.
df1 = pd.DataFrame(data1)
print("\nthe first dataframe: \n", df1)

df2 = pd.DataFrame(data2)
print("\nthe second dataframe: \n", df2)

## join the two dataframe in new dataframe.
dfresult = df1.join(df2)
print("\nthe dataframe after join:\n", dfresult)  ## join two dataframe using join.
```

# #9. COMBINE TWO PANDAS SERIES-

to combine two pandas series into one in python, use the **combine( )** method. it is used a specific function for the decision, mentionby the user as a parameter of the combine() method.

```python
import pandas as pd

#data 
data1 = [10, 30, 60, 90, 120]
data2 = [60, 40, 50, 80, 100]

## create a series using the series( ) method.
s1 = pd.Series(data1)
s2 = pd.Series(data2)

## print both the series.
print("series1: \n" , s1)
print("series2: \n" , s2)

## function. check  which element is bigger and print the element in the combine series.
def mixseries(s1,s2):
    if s1 > s2:
        return s1
    else :
        return s2

## combine the two series in one. and the function return the largest value. from series by combining both series.
result = s1.combine(s2, mixseries)

## displying the series.
print("after combine : \n", result)
```

# #11. PANDAS WORKING WITH CATEGORIES -
in this, we will understand how to work with Categories in pandas.
* Append Category - use the **add_categories( )** method. to append category.
* Remove category - use the **remove_categories( )** method. to remove category.

## Q.1 Append new categories.

```python
import pandas as pd

## create a categorical series.
s = pd.Series(["a", "e", "i", "o", "u"], dtype = "category")

## Display the series.
print("\nSeries before append = \n", s)

## append category.
s = s.cat.add_categories("A")

## Display the series.
print("\nSeries after append category = \n", s)

## append categories.
s = s.cat.add_categories(["E", "I", "O", "U"])

## Display the series.
print("\nSeries after append categories = \n", s)
```

## Q.2 Remove a category

```python
import pandas as pd

## create a categorical series.
s = pd.Series(["a", "e", "i", "o", "u"], dtype = "category")

## Display the series.
print("\nSeries before append = \n", s)

## remove category.
s = s.cat.remove_categories("a")

## Display the series.
print("\nSeries after remove category = \n", s)

## remove categories.
s = s.cat.remove_categories(["e", "o"])

## Display the series.
print("\nSeries after remove categories = \n", s)
```

# #.16 PANDAS ADD A NEW COLUMN -
in this we will insert a column to an array created DataFrame. using 2 methods - 
1. insert ( ) -> add new column at specific loction.(set the location) ; **syntax -> df.insert(2 , "column name", [columns data])** --> df.insert(loction where insert, name of columns, values of new column.)
     * Works in place  directly modifies the existing DataFrame.
     * Does not return a new DataFrame  it returns None.


2. assign ( ) -> add new column at the end defult. ; **syntax -> result =  df.assign( column name = [columns data])** --> df.insert(name of columns, values of new column.)

# #.25 PANDAS GROUP THE DATA -

## Q.1 split the object and combine the result. hint( used groupby( ) method in it.)

```python
import pandas as pd

## data
data = {
    "Players" : [ "AMIT", "AMIT", "RAGHAV", "ARPIT", "SURAJ", "JAGGI",],
    "Rank" : [1, 4, 3, 5, 2, 7],
    "Year" : [2023, 2019, 2021, 2022, 2018, 2019]
}

## dataframe create and display.
df = pd.DataFrame(data)
print("\n the data is: \n", df)

## group the data on player value.
resdf = df.groupby('Players')
print("\n the data grouped by players: \n",resdf.first())    ## using resdf.first() make the players as index of the dataframe.

## group the data on year value.
resdf = df.groupby('Year')
print("\n the data grouped by Years: \n",resdf.first())    ## using resdf.first() make the  as index of the dataframe.
```

## Q.1 plot a dataframe in pandas. 

```python
import pandas as pd
import matplotlib.pyplot as plt

## data
data = {
    "temp" : [18, 25, 35, 38, 49, 28, 54, 30, 27, 10],
    "humidity": [32, 45, 54, 63, 72, 28, 24, 27, 48, 35],
    "wind" : [12, 32, 54, 64, 23, 54, 73, 83, 23, 43]
}

## create dataframe
df = pd.DataFrame(data)

## plot the dataframe
df.plot()

## display the graph
plt.grid()
plt.show()
```

## Q.2 plot HISTOGRAM. is a graphical representation of show frequency distribution. create histogram set the kind arugment of the plot. i.g => df['wind'].plot(kind='hist')

```python
import pandas as pd
import matplotlib.pyplot as plt

## data
data = {
    "temp" : [18, 25, 35, 38, 49, 28, 54, 30, 27, 10],
    "humidity": [32, 45, 54, 63, 72, 28, 24, 27, 48, 35],
    "wind" : [12, 32, 54, 64, 23, 54, 73, 83, 23, 43]
}

## create dataframe
df = pd.DataFrame(data)

## plot the histrogram.
df["wind"].plot(kind='hist')

## display
plt.show()
```

## Q.3 plot PIE CHART. is a pictorial form, divided into slice. use plot.pie( ) to draw.

```python
import pandas as pd
import matplotlib.pyplot as plt

## data
data = {
    "temp" : [18, 25, 35, 38, 49, 28, 44, 30, 27, 10],
    "humidity": [32, 45, 54, 63, 72, 28, 24, 27, 48, 35],
    "wind" : [12, 32, 54, 64, 23, 54, 73, 83, 23, 43]
}

## create dataframe
df = pd.DataFrame(data, index = ["manali","dehradun","haridwor","meerut","delhi","chennai","noida","pune","gohatti","leh"])

## plot the piechart.
df.plot.pie(y="temp")

## display the chart
plt.show()
```

## Q.4 plot SCATTER PLOT. is a chart display the realationship between two variables. each value in dataset. it will create scatter plot, set kind argument. we will also set x,y axis i.g -> df.plot(kind='scatter',x='temp', y='humidty')

```python
import pandas as pd
import matplotlib.pyplot as plt

## data
data = {
    "temp" : [18, 25, 35, 38, 49, 28, 44, 30, 27, 10],
    "humidity": [32, 45, 54, 63, 72, 28, 24, 27, 48, 35],
    "wind" : [12, 32, 54, 64, 23, 54, 73, 83, 23, 43]
}

## create dataframe
df = pd.DataFrame(data, index = ["manali","dehradun","haridwor","meerut","delhi","chennai","noida","pune","gohatti","leh"])

## plot the scatter plot.
df.plot(kind = 'scatter', x = 'temp', y = 'humidity')

## display the chart
plt.show()
```
